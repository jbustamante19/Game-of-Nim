```python

import numpy as np
import random

#Class pile is rather simple in that it randomly selects the number of marbles the pile will consist of.
#It also hosts the important updateMarbleNum method which will be called by methods within the player class.
class pile:
    def __init__(self):
        n = random.randint(10,100)
        self.marbleNum = n
        print('')
        print("Marbles in the pile: "+str(self.marbleNum))
        print('')
    def updateMarbleNum(self,take):
        self.marbleNum-=take

        
#In this class we instantiate the player objects which will be the computer and human user. 


class player:
    
    def __init__(self,name):
        self.name = name
        self.take = 0
        
    #This method is called from the game class whenever it is the users turn to remove a marble from the pile. 
    #It enforces the rule that the user can only take from one but up to half of the number of marbles in the pile. 
    #It then calls the updateMarbleNum method from pile class to update the number of marbles in the pile after the 
    #user has specified how many marbles to take.
        
    def humanMove(self,pile1):
        self.take = int(input("Take between 1 and "+str(int(pile1.marbleNum/2.0))+" marbles: "))
        while self.take == 0 or self.take> int(pile1.marbleNum/2.0) :
            self.take = int(input("Must take between 1 and "+str(int(pile1.marbleNum/2.0))+" marbles: "))
            
        pile1.updateMarbleNum(self.take)
        
        print(self.name+" took: "+str(self.take))
        print("Marbles Left: "+str(pile1.marbleNum))
    #In the computerMove method within the player class there exists two modes. The dumb mode shown here 
    #shows the move the computer makes when it leaves the number of marbles to remove to chance. 
    #After the number is specified the pile is adjusted via the pile class. 
    
    def computerMove(self,pile1):
        mode=random.randint(0,1)
        if mode == 0:
            print("Computer Move in Dumb Mode")
            self.take = random.randint(1,int((pile1.marbleNum)/2.0)) 
            pile1.updateMarbleNum(self.take)
            print("Computer took: "+str(self.take))
            print("Marbles Left: "+str(pile1.marbleNum)) 
            
        #The smart mode shown here specifies that when the number of marbles left in the pile 
        #is not a power of two minus 1, the number of marbles removed should leave the pile with
        #such quantity of marbles. The pile of marbles is of course updated after each iteration.

        if mode ==1:
            print("Computer Move in Smart Mode")
            if (pile1.marbleNum == 3 or pile1.marbleNum == 7 or pile1.marbleNum == 15 
                or pile1.marbleNum == 31 or pile1.marbleNum == 63 or pile1.marbleNum<3):
                self.take = random.randint(1,int((pile1.marbleNum)/2.0)) 
            else:
                turnMarbleNum = 1000000
                while (turnMarbleNum != 3 and turnMarbleNum != 7 and turnMarbleNum != 15 
                       and turnMarbleNum != 31 and turnMarbleNum != 63):
                    self.take = random.randint(1,int((pile1.marbleNum)/2.0))
                    turnMarbleNum = pile1.marbleNum-self.take    
            pile1.updateMarbleNum(self.take)
            print("Computer took: "+str(self.take))
            print("Marbles left: "+str(pile1.marbleNum))
                  
class game:
    
    def __init__(self,player1):
        
        #class is used as the main class and is used to instantiate objects of the other two classes. 
        
        self.player1 = player(player1)
        self.player2 = player('Computer')
        self.pile1 = pile()
        
        #Within the game class we go on to randomly select if whether the user or computer take the 
        #first move. We accomplish this by using a random number generator that returns either a zero or one. 
        
        self.turn = random.randint(0,1)
        if self.turn == 0:
            self.player1.humanMove(self.pile1)
            self.turn = 0
        if self.turn == 1: 
            self.player2.computerMove(self.pile1)
            self.turn = 1
        #I make use of a while loop in order to allow for the iteration of turns between the computer 
        #and the human opponent. I also call the player class to deal with the appropriate movements. 
        #Lastly, I set winning and losing conditions.
           
        while self.pile1.marbleNum>0:
            self.turn+=1
            if self.turn%2 == 0 :
                if self.pile1.marbleNum == 1:

                    print("LOSERR!")
                    break 
                self.player1.humanMove(self.pile1)
                
            if self.turn%2 != 0 :
                if self.pile1.marbleNum == 1:
                    print("YOU WIN!")
                    break
                self.player2.computerMove(self.pile1)
            print('')
                
game1 = game(input("Enter Player Name: "))      
                
```
